```python
import numpy as np
import matplotlib.pyplot as plt
import requests
from io import BytesIO
from PIL import Image
```

```python
response = requests.get('http://vignette2.wikia.nocookie.net/grayscale/images/4/47/Lion.png/revision/latest?cb=20130926182831')
load_lion_arr = Image.open(BytesIO(response.content))
lion_arr = np.array(load_lion_arr)
```

```python
plt.imshow(lion_arr)
```

# Convolution with 2-D Kernels

## 평균 블러링(Average Blurring)
앞서 설명했듯이 블러링은 초점이 맞지 않듯이 영상을 흐릿하게 하는 작업을 뜻합니다. 가장 간단한 블러링 방법으로는 평균 블러링이 있습니다. 평균 블러링은 주변 픽셀 값들의 평균을 적용합니다. 주변 픽셀들의 평균값을 적용하면 픽셀 간 차이가 적어져 선명도가 떨어져 전체적으로 흐릿해집니다.



```python
import cv2
from google.colab.patches import cv2_imshow

# 5x5 평균 필터 커널 생성  ---②
kernel = np.ones((5,5))/5**2
kernel2 = np.ones((3,3))/5**2
kernel3 = np.ones((10,10))/5**2
kernel4 = np.ones((20,20))/5**2

'''
#5x5 평균 필터 커널 생성    ---①
kernel = np.array([[0.04, 0.04, 0.04, 0.04, 0.04],
                   [0.04, 0.04, 0.04, 0.04, 0.04],
                   [0.04, 0.04, 0.04, 0.04, 0.04],
                   [0.04, 0.04, 0.04, 0.04, 0.04],
                   [0.04, 0.04, 0.04, 0.04, 0.04]])
'''

# 필터 적용             ---③
blured = cv2.filter2D(lion_arr, -1, kernel)
blured2 = cv2.filter2D(lion_arr, -1, kernel2)
blured3 = cv2.filter2D(lion_arr, -1, kernel3)
blured4 = cv2.filter2D(lion_arr, -1, kernel4)

# 결과 출력
cv2_imshow(lion_arr)
cv2_imshow(blured) 
cv2_imshow(blured2) 
cv2_imshow(blured3)
cv2_imshow(blured4) 
```

# 엣지 필터 (Edge filter)

```python


edge_kernel_1 = np.array([[1, 0, -1],
                          [0, 0, 0],
                          [-1, 0, 1]])

edge_kernel_2 = np.array([[0, 1, 0],
                          [1, -4, 1],
                          [0, 1, 0]])

edge_kernel_3 = np.array([[-1, -1, -1],
                          [-1, 8, -1],
                          [-1, -1, -1]])


# 필터 적용           
edge_1 = cv2.filter2D(lion_arr, -1, edge_kernel_1)
edge_2 = cv2.filter2D(lion_arr, -1, edge_kernel_2)
edge_3 = cv2.filter2D(lion_arr, -1, edge_kernel_3)

# 결과 출력
cv2_imshow(lion_arr)
cv2_imshow(edge_1) 
cv2_imshow(edge_2) 
cv2_imshow(edge_3) 
```

```python
plt.imshow(edge_kernel_3)
```

# 샤프 필터 (sharpen filter)

```python
sharpen_kernel = np.array([[0, -1, 0],
                           [-1, 5, -1],
                           [0, -1, 0]])

unsharp_kernel = np.array([[1, 4, 6, 4, 1],
                           [4, 16, 24, 16, 4],
                           [6, 24, -476, 24, 6],
                           [4, 16, 24, 16, 4],
                           [1, 4, 6, 4, 1]]) / -256

# 필터 적용           
sharpen_img = cv2.filter2D(lion_arr, -1, sharpen_kernel)
unsharp_img = cv2.filter2D(lion_arr, -1, unsharp_kernel)

# 결과 출력
cv2_imshow(lion_arr)
cv2_imshow(sharpen_img) 
cv2_imshow(unsharp_img) 
```
